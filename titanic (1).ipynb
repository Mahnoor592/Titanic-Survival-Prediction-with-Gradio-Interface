# **DATA CLEANING**

Filling the missing NAN Values

```python
import numpy as np
mean_age = np.mean(df.Age)
print(mean_age)
```

```python
df["Age"] = df["Age"].fillna(mean_age)
```

```python
df.isnull().sum()
```

```python
df.drop(["Cabin"], axis = 1, inplace = True)
# drop the column Cabin because it has too many missing values
# When a column has too many missing values, filling it would require guessing or assumptions, which may reduce accuracy.
# axis =  0 -> means row
# axis = 1 -> means column
# inplace = true -> make directly changes in the original df
```

```python
df.isnull().sum()
```

```python
df["Fare"] = df["Fare"].fillna(df["Fare"].mean())
```

```python
df.isnull().sum()
```

In case of Nan Values in Embarked Colmun

```python
df.Embarked.unique()
```

```python
# As these values are characters not numbers to fill the Nan values we will have to fill it with most common category
mode_embarked = df.Embarked.mode()
print(mode_embarked)
# df.Embarked -> access the Embarked column
# .mode() -> return the most frequent

# 0 -> index of S , as in output we can get multiple values that appeared equal times
```

```python
mode_embarked[0]
```

```python
df["Embarked"] = df["Embarked"].fillna(mode_embarked[0])
```

```python
df.isnull().sum()
```

```python
df.info()
```

Dropping name column because with it we can not predict if the person survive

```python
df.drop(["Name"], axis = 1, inplace = True)
```

```python
df.info()
```

As gender can effect the survival rate


```python
df["Sex"] = df["Sex"].replace("male", 1)
df["Sex"] = df["Sex"].replace("female", 0)
# replacing male with 1 and female with 0
```

```python
df.info()
```

Dropping Passenger Id and Ticket as well

```python
df.drop(["PassengerId"], axis = 1, inplace = True)
df.drop(["Ticket"], axis = 1, inplace = True)
```

```python
df.info()
```

# **One-Hot-Encoding**

Now Converting the data to Numeric data because computer cannot understand it
using one hot encoding

```python
# #Import tools for transforming categorical data into machine readable format

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
```

```python
transformer = make_column_transformer(
    (OneHotEncoder(), ["Sex","Embarked"]),
    remainder = "passthrough"
)
# apply one hot encoding to Sex and Embarked
# remainder = passthrough -> keeps all other columns unchanged

transformed = transformer.fit_transform(df)
# Learns the categories and transforms them into new binary (0/1) columns.
# the result is Numpy array
```

Convert back to dataframe

```python
df = pd.DataFrame(
    transformed,
    columns = transformer.get_feature_names_out()
)
# Converts the transformed NumPy array back to a pandas DataFrame.
# Assigns new column names like onehotencoder__Sex_female, onehotencoder__Embarked_S, etc.
```

# **Training and Testing**

```python
# Import library for splitting into training ans testing
from sklearn.model_selection import train_test_split
```

```python
# 80% -> training
x_train, x_test, y_train,y_test = train_test_split(x, y, test_size = 0.2)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Initialize Model
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train) # Model is Trained
```

```python
# predicting the survival
y_pred = model.predict(x_train)
```

```python
import gradio as gr

from sklearn.ensemble import RandomForestClassifier

# Assuming model and transformer are already trained
# model = RandomForestClassifier()
# model.fit(x_train, y_train)

# Prediction function
def predict_survival(Pclass, Sex, Age, SibSp, Parch, Fare, Embarked):
    Sex = 1 if Sex.lower() == "male" else 0

    input_df = pd.DataFrame([{
        "Sex": Sex,
        "Embarked": Embarked,
        "Pclass": Pclass,
        "Age": Age,
        "SibSp": SibSp,
        "Parch": Parch,
        "Fare": Fare,
        "Survived": 0  # dummy column
    }])

    input_transformed = transformer.transform(input_df)
    input_transformed = input_transformed[:, :-1]  # drop dummy Survived

    prediction = model.predict(input_transformed)
    return "🟢 Survived" if prediction[0] == 1 else "🔴 Not Survived"

# Build Gradio Interface
with gr.Blocks() as iface:
    gr.Markdown("# 🚢 Titanic Survival Prediction")
    gr.Markdown("Fill the passenger details to predict survival.")

    with gr.Row():
        Pclass = gr.Dropdown([1, 2, 3], label="Passenger Class")
        Sex = gr.Radio(["male", "female"], label="Sex")
        Age = gr.Number(label="Age", value=30)

    with gr.Row():
        SibSp = gr.Number(label="Siblings Aboard", value=0)
        Parch = gr.Number(label="Parents/Children Aboard", value=0)
        Fare = gr.Number(label="Fare", value=30.0)
        Embarked = gr.Dropdown(["C", "Q", "S"], label="Port of Embarkation")

    predict_btn = gr.Button("Predict Survival")
    output = gr.Textbox(label="Prediction Result")

    predict_btn.click(fn=predict_survival,
                      inputs=[Pclass, Sex, Age, SibSp, Parch, Fare, Embarked],
                      outputs=output)

# Launch the app
iface.launch(share=True)

```
